```python
import matplotlib.pyplot as plt
import matplotlib as mpl
import numpy as np
```

```python
mpl.rcParams["font.family"] ='serif'
mpl.rcParams["font.size"] =7
```

```python
ann = np.array([97.5, 96.8, 96.5])
cnn = np.array([99.5, 99.32, 99.5])
acc1 = np.array(['Training', 'Validation', 'Test'])
acc = np.array([1,1.5,2])
width = 0.2
```

```python
fig, ax = plt.subplots(1,figsize = (4,3),dpi=200)
# fig.size = (4,3)
ax.bar(acc- width/2,ann, width/1.2,label = 'CNN')
ax.bar(acc+ width/2,cnn, width/1.2, label = 'ANN')
ax.set_xticks(acc)
ax.set_ylim(0,119.9)
ax.set_xticklabels(acc1)
ax.set_ylabel('Accuracy (%)')
ax.legend(loc='upper left')
plt.show()
```
